```python
import plotly.graph_objects as go
import datetime
import numpy as np
import pandas as pd
np.random.seed(1)
```

```python
nodes = pd.read_csv('../dataset/pet_supplies.csv')
edges = pd.read_csv('../dataset/pet_supplies_edges.csv', skipinitialspace=True)
```

```python
categories = nodes[["name"]]
categories.shape
# ids = nodes[["id"]]
# categories.values.reshape(1, 555)
# categories.values.ravel()
# categories.values.ravel()[0:categories_to_plot-1]
```

```python
total_categories = nodes.shape[0]
categories_to_plot = 50
adj_matrix = np.ones(shape=(categories_to_plot, categories_to_plot), dtype=int)
```

```python
for _, row in edges.iterrows():
    if row.src < categories_to_plot and row.dest < categories_to_plot:
        adj_matrix[row.src,row.dest]=0
```

```python
# fig = go.Figure(data=go.Heatmap(
#         z=adj_matrix,
#         x=np.arange(1,total_categories),
#         y=np.arange(1,total_categories),
# ))
fig = go.Figure(data=go.Heatmap(
        z=adj_matrix,
        x=categories.values.ravel()[0:categories_to_plot-1],
        y=np.arange(1,categories_to_plot),
        colorscale='Blackbody'
))

fig.update_layout(
    title='Adjacency Heatmap',
    yaxis_nticks=50,
    xaxis_nticks=100)

fig.show()
```
